```python
import pandas as pd
from bertopic import BERTopic

# Convert the 'date' column to a list of timestamps
timestamps = df_r['date'].to_list()

# Convert the 'cleaned_tokens' column to a list of tokenized tweets
tokenized_tweets = df_r['cleaned_tokens'].to_list()

# Flatten the list of lists to make it compatible with BERTopic
flattened_tweets = [token for sublist in tokenized_tweets for token in sublist]

# Initialize the BERTopic model with verbosity enabled
topic_model = BERTopic(verbose=True)

# Fit the model to the tokenized tweets and transform the data to get topics and probabilities
topics, probs = topic_model.fit_transform(flattened_tweets)

# Analyze topics over time
topics_over_time = topic_model.topics_over_time(flattened_tweets, timestamps, nr_bins=20)

# Optional: Display or save the result
print(topics_over_time)
```
