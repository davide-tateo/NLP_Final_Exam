```python
%pip install bertopic
import pandas as pd
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

# Load the data
data = pd.read_csv('df_r.csv')

# Ensure 'cleaned_tokens' are evaluated to list of tokens
data['cleaned_tokens'] = data['cleaned_tokens'].apply(eval)

# Convert cleaned tokens back to strings for BERTopic
data['text'] = data['cleaned_tokens'].apply(lambda tokens: ' '.join(tokens))

# Extract the texts and corresponding months
texts = data['text'].tolist()
months = data['month'].tolist()

# Create and fit the BERTopic model
topic_model = BERTopic()
topics, probabilities = topic_model.fit_transform(texts)

# Prepare data for topics over time
timestamps = data['month']

# Generate topics over time
topics_over_time = topic_model.topics_over_time(texts, topics, timestamps, nr_bins=4)

# Plot the topic evolution
topic_model.visualize_topics_over_time(topics_over_time)
```
